### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Create a bar plot
plt.figure(figsize=(6, 4))

# Error handling: Ensure categories, values, and colors have the same length
assert len(categories) == len(values) == len(colors), "categories, values, and colors must have the same length"

plt.bar(categories, values, color=colors)


# Labeling the axes and the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Colors')

# Show the plot
plt.show()
```
