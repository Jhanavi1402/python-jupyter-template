### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'Javascript']
sizes = [40, 25, 20, 15]

# Error Handling: Ensure labels and sizes have the same length
assert len(labels) == len(sizes), "Error: labels and sizes must have the same length"

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Title
plt.title('Pie Chart with Percentages')

# Show the plot
plt.show()
```
